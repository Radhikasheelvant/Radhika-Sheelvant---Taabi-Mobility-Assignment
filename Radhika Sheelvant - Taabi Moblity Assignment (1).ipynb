```python
#handling missing values and any irregularities in the data.

missing_values = data.isna().sum()
```

```python
#drop columns with missing values

clean_data = data.dropna(axis=1, how='any')
```

# Data cleaning and preprocessing:

```python
clean_data.head()
```

```python
for col in clean_data.columns:
    if clean_data[col].isna().sum() > 0:
        if clean_data[col].dtype == pd.number:
            clean_data[col].fillna(clean_data[col].mean(), inplace=True)
    
print(clean_data.head())
```

```python
clean_data.head()
```

```python
#Converting categorical data into numerical data

#import labelencder
from sklearn.preprocessing import LabelEncoder

#initialize LabelEncoder
label_encoder = LabelEncoder()

# repeating through each column and encode if dtype is object
for col in clean_data.columns:
    if clean_data[col].dtype == 'object':
        clean_data[col] = label_encoder.fit_transform(clean_data[col])

#print("\nDataFrame after Label Encoding with numpy array:")
#print(clean_data)

# One-Hot Encoding with numpy array
# Use pd.get_dummies after converting numpy array to DataFrame
df_encoded = pd.get_dummies(clean_data)

print("\nDataFrame after One-Hot Encoding with numpy array:")
print(df_encoded.head())
```
